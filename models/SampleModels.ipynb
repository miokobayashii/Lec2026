# SIRモデル (感染症の流行過程, Susceptible, Infected, Recovered)

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 微分方程式の定義（戻り値をNumPy配列にして計算しやすくする）
def sir_model(y, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I
    dIdt = beta * S * I - gamma * I
    dRdt = gamma * I
    return np.array([dSdt, dIdt, dRdt])  # 配列で返すのがコツ！

# 2. 初期設定
S0, I0, R0 = 0.99, 0.01, 0.00
y = np.array([S0, I0, R0])

beta = 0.3
gamma = 0.1

# 3. 時間軸の設定と dt の計算
t_end = 160
dt = 0.1  # 時間刻み（ステップサイズ）
t_steps = np.arange(0, t_end, dt)  # 0から160まで0.1刻みの配列

# データ格納用のリスト（初期値も最初に入れておく）
S = [y[0]]
I = [y[1]]
R = [y[2]]

# 4. オイラー法によるループ計算
for _ in t_steps[1:]:  # 初期値の次は2ステップ目から計算
    # y_next = y + (dy/dt) * dt
    next_y = y + sir_model(y, beta, gamma) * dt
    y = next_y

    # データを記録
    S.append(y[0])
    I.append(y[1])
    R.append(y[2])

# 5. 可視化
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({
    "font.size": 20,
    "axes.labelsize": 20,
    "axes.titlesize": 20,
    "xtick.labelsize": 20,
    "ytick.labelsize": 20,
    "xtick.direction": "in",
    "ytick.direction": "in"
})

plt.figure(figsize=(10, 6), dpi=100)
plt.plot(t_steps, S, label='Susceptible', color='blue', lw=2)
plt.plot(t_steps, I, label='Infected', color='red', lw=2)
plt.plot(t_steps, R, label='Recovered', color='green', lw=2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title('SIR Model Simulation',fontweight='bold')
plt.xlabel('Days')
plt.ylabel('Proportion of Population')
plt.show()
```

# Lotka-Volterra predator-prey モデル (捕食-被食者の個体数の変動）

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. ロトカ・ヴォルテラモデルの微分方程式の定義
def lotka_volterra(y, alpha, beta, delta, gamma):
    X, Y = y  # X: 被食者（ウサギ）, Y: 捕食者（オオカミ）

    # 微分方程式
    dXdt = alpha * X - beta * X * Y
    dYdt = delta * X * Y - gamma * Y

    return np.array([dXdt, dYdt])

# 2. 初期状態とパラメータの設定
X0 = 40.0  # ウサギの初期個体数
Y0 = 9.0   # オオカミの初期個体数
y = np.array([X0, Y0])

# 生態系パラメータ
alpha = 1.0  # ウサギの固有増殖率
beta = 0.1   # オオカミによるウサギの捕食率
delta = 0.03 # ウサギを食べたことによるオオカミの増殖率
gamma = 0.75 # オオカミの自然死亡率

# 3. 時間軸の設定（オイラー法用）
t_end = 50
dt = 0.01  # 時間刻み（細かくすることでオイラー法の誤差を抑える）
t_steps = np.arange(0, t_end, dt)

# データ格納用のリスト（初期値をあらかじめ格納）
X_history = [y[0]]
Y_history = [y[1]]

# 4. オイラー法によるループ計算
for _ in t_steps[1:]:
    # 次の状態 ＝ 現在の状態 ＋ 傾き(微分) × dt
    next_y = y + lotka_volterra(y, alpha, beta, delta, gamma) * dt
    y = next_y

    # データを記録
    X_history.append(y[0])
    Y_history.append(y[1])

# 5. 可視化（2つの視点）
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), dpi=100)
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({
    "font.size": 20,
    "axes.labelsize": 20,
    "axes.titlesize": 20,
    "xtick.labelsize": 20,
    "ytick.labelsize": 20,
    "xtick.direction": "in",
    "ytick.direction": "in"
})
# 視点①：時間波形（Time Series）
max_population = max(max(X_history), max(Y_history))
y_limit_upper = max_population * 1.35

ax1.plot(t_steps, X_history, label='Prey (Rabbits)', color='blue', lw=2)
ax1.plot(t_steps, Y_history, label='Predator (Wolves)', color='red', lw=2)
ax1.set_title('Time Series Dynamic',  fontweight='bold')
ax1.set_xlabel('Time')
ax1.set_ylim(0,y_limit_upper)          # 縦軸の下限を0、上限をゆとりを持たせた値に固定
ax1.legend(loc='upper right')
ax1.set_ylabel('Population')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(loc='upper right')

# 視点②：相平面（Phase Plane）★本講義の核心★
ax2.plot(X_history, Y_history, color='purple', lw=2)
ax2.scatter(X0, Y0, color='green', s=100, zorder=5, label='Start point')
ax2.set_title('Phase Plane Topology', fontweight='bold')
ax2.set_xlabel('Prey Population (X)')
ax2.set_ylabel('Predator Population (Y)')

ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend(loc='upper right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. オレゴネータ（BZ反応モデル）の微分方程式の定義
def oregonator(y, epsilon, q, f):
    x, z = y  # x: 臭素酸イオンなどの過渡物質（高速）, z: 触媒の酸化状態（低速）

    # 無次元化された微分方程式
    #dxdt = (x + q*x - x**2 - 2*q*x - x*z + q*z) / epsilon  # 非常に速い変化
    #dzdt = x - z                                          # 緩やかな変化

    # 上記は古典的な簡易オレゴネータの表現形式（FHNのvとwに相当）
    # より直感的な表現：
    dxdt = (x * (1 - x) - f * z * (x - q) / (x + q)) / epsilon
    dzdt = x - z

    return np.array([dxdt, dzdt])

# 2. 初期状態とパラメータの設定（BZ反応が綺麗に振動する定番値）
x0 = 0.1
z0 = 0.1
y = np.array([x0, z0])

# パラメータ
epsilon = 0.02  # 💡FHNの φ 相当。時間尺度の分離（xが超高速、zが低速）を生む
q = 0.002       # 化学反応の微小定数
f = 1.0         # 化学量論係数（ブレーキの強さ。これを動かすと分岐が起きる）

# 3. 時間軸の設定（超高速な変化を捉えるため dt は極小にする）
t_end = 30
dt = 0.0005     # オイラー法の誤差と発散を防ぐための極小ステップ
t_steps = np.arange(0, t_end, dt)

# データ格納用
x_history = [y[0]]
z_history = [y[1]]

# 4. オイラー法によるループ計算
for _ in t_steps[1:]:
    next_y = y + oregonator(y, epsilon, q, f) * dt

    # 数値計算の安定化（濃度がマイナスにならないようセーフティを入れる）
    next_y = np.maximum(next_y, 0.0)

    y = next_y
    x_history.append(y[0])
    z_history.append(y[1])

# 5. 可視化
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), dpi=100)

# 視点①：時間波形
max_conc = max(max(x_history), max(z_history))
y_limit_upper = max_conc * 1.35
ax1.plot(t_steps, x_history, label='x (Activator)', color='cyan', lw=2)
ax1.plot(t_steps, z_history, label='z (Inhibitor)', color='magenta', lw=2)
ax1.set_xlabel('Time')
ax1.set_ylabel('Concentration')
ax1.set_ylim(-0.1,y_limit_upper)
ax1.legend(loc='upper right')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.set_title('BZ Reaction Temporal Oscillation', fontweight='bold')

# 視点②：相平面
ax2.plot(x_history, z_history, color='purple', lw=2)
ax2.scatter(x0, z0, color='green', s=100, label='Start Point')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.legend(loc='upper right')
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.set_title('BZ Reaction Limit Cycle', fontweight='bold')

plt.tight_layout()
plt.show()
```
